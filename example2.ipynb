# Examples
### 1
Apply a production to a graph.

```python
from hypergrammar.productions.prod_2 import Prod2
from hypergrammar.hypergraph import Hypergraph

def run_prod2(hg: Hypergraph):
    prod2 = Prod2()
    new_hg = prod2.apply(hg)
    if new_hg:
        print(new_hg.get_edges())
        new_hg.draw(use_positional_parameters=True)
    else:
        print("Production could not be applied.")
    return new_hg
```

```python
from hypergrammar.edge import Edge, EdgeType

hg = Hypergraph()
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "B"}), {"R": 1, "B":0}))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "C"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "A"})))

hg.set_vertex_parameter("A", {"x": 1, "y": 1, "z": 1})
hg.set_vertex_parameter("B", {"x": 2, "y": 2.1, "z": 2})
hg.set_vertex_parameter("C", {"x": 3, "y": 3, "z": 3})

# I guess nie mozemy wystartowa z wierzchokami ktre maj takie wsprzdne ze si przecinaj, trzeba to zmieni dopiero na kocu

# hg.set_vertex_parameter("A", {"x": 0, "y": 2, "z": 0})
# hg.set_vertex_parameter("B", {"x": 0, "y": 0, "z": 0})
# hg.set_vertex_parameter("C", {"x": 1, "y": 1, "z": 0})

hg.draw(use_positional_parameters=True)
```

```python
run_prod2(hg)
```

```python
hg = Hypergraph()

# HORIZONTAL EDGES
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "B"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "C"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "D"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"E", "F"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"G", "H"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"H", "I"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"I", "J"})))


#VERTICAL EDGES
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "G"})))

hg.add_edge(Edge(EdgeType.E, frozenset({"B", "H"}), {"R": 1, "B": 0}))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "E"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"E", "H"})))

hg.add_edge(Edge(EdgeType.E, frozenset({"C", "I"})))

hg.add_edge(Edge(EdgeType.E, frozenset({"D", "F"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"F", "J"})))


hg.set_vertex_parameter("A", {"x": 0, "y": 0, "z": 0})
hg.set_vertex_parameter("B", {"x": 2, "y": 0, "z": 0})
hg.set_vertex_parameter("C", {"x": 3, "y": 0, "z": 0})
hg.set_vertex_parameter("D", {"x": 4, "y": 0, "z": 0})

hg.set_vertex_parameter("E", {"x": 2.1, "y": 1, "z": 0})
hg.set_vertex_parameter("F", {"x": 4, "y": 1, "z": 0})

hg.set_vertex_parameter("G", {"x": 0, "y": 2, "z": 0})
hg.set_vertex_parameter("H", {"x": 2, "y": 2, "z": 0})
hg.set_vertex_parameter("I", {"x": 3, "y": 2, "z": 0})
hg.set_vertex_parameter("J", {"x": 4, "y": 2, "z": 0})

hg.draw(use_positional_parameters=True)
```

```python
run_prod2(hg)
```
