# Examples
### 1
Apply a production to a graph.

```python
from hypergrammar.hypergraph import Hypergraph
from hypergrammar.edge import Edge, EdgeType
from hypergrammar.productions.prod_5 import Prod5

# Vertices: A, B, C, D; intermediates: X, Y, Z, W
# Each E edge is 'broken' via an intermediate
hg = Hypergraph()
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "X"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"X", "B"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "Y"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"Y", "C"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "Z"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"Z", "D"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"D", "W"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"W", "A"})))
hg.add_edge(Edge(EdgeType.Q, frozenset({"A", "B", "C", "D"}), {"R": 1}))

# Optionally set coordinates for visualization
hg.set_vertex_parameter("A", {"x": 0, "y": 0})
hg.set_vertex_parameter("B", {"x": 1, "y": 0})
hg.set_vertex_parameter("C", {"x": 1, "y": 1})
hg.set_vertex_parameter("D", {"x": 0, "y": 1})
hg.set_vertex_parameter("X", {"x": 0.5, "y": 0})
hg.set_vertex_parameter("Y", {"x": 1, "y": 0.5})
hg.set_vertex_parameter("Z", {"x": 0.5, "y": 1})
hg.set_vertex_parameter("W", {"x": 0, "y": 0.5})

print("Before applying Prod5:")
hg.draw(use_positional_parameters=True)
```

```python
prod5 = Prod5()
new_hg = prod5.apply(hg)

print("After applying Prod5:")
if new_hg:
    new_hg.draw(use_positional_parameters=True)
else:
    print("Production could not be applied.")
```

```python
from hypergrammar.hypergraph import Hypergraph
from hypergrammar.edge import Edge, EdgeType
from hypergrammar.productions.prod_1 import Prod1

hg = Hypergraph()
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "B"}), {"B": 1, "R": 0}))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "C"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "D"})))
hg.add_edge(Edge(EdgeType.E, frozenset({"D", "A"})))
hg.add_edge(Edge(EdgeType.Q, frozenset({"A", "B", "C", "D"}), {"R": 1}))

hg.set_vertex_parameter("A", {"x": 0, "y": 0})
hg.set_vertex_parameter("B", {"x": 1, "y": 0})
hg.set_vertex_parameter("C", {"x": 1, "y": 1})
hg.set_vertex_parameter("D", {"x": 0, "y": 1})

print("Before applying Prod1:")
hg.draw(use_positional_parameters=True)
```

```python
prod1 = Prod1()
new_hg = prod1.apply(hg)
if new_hg:
    print(new_hg.get_edges())
    new_hg.draw(use_positional_parameters=True)
else:
    print("Production could not be applied.")
```
