```python
import os
import sys

proj_root = os.path.abspath(os.path.join(os.path.abspath(""), ".."))
sys.path.append(str(proj_root))
from hypergrammar.hypergraph import Hypergraph
from hypergrammar.edge import Edge, EdgeType
from hypergrammar.productions.prod_4 import Prod4


hg = Hypergraph()
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "B"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "C"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "D"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"D", "A"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.Q, frozenset({"A", "B", "C", "D"}), {"R": 0}))

hg.set_vertex_parameter("A", {"x": 0, "y": 0})
hg.set_vertex_parameter("B", {"x": 1, "y": 0})
hg.set_vertex_parameter("C", {"x": 1, "y": 1})
hg.set_vertex_parameter("D", {"x": 0, "y": 1})

hg.draw(use_positional_parameters=True)
```

```python

prod0 = Prod4()
new_hg = prod0.apply(hg)
if new_hg:
    print(new_hg.get_edges())
    new_hg.draw(use_positional_parameters=True)
else:
    print("Production could not be applied.")
```

```python
from hypergrammar.hypergraph import Hypergraph
from hypergrammar.edge import Edge, EdgeType
from hypergrammar.productions.prod_0 import Prod0
from typing import Any, Mapping, Optional, List
from shapely.geometry import Point, MultiPoint

hg = Hypergraph()

# --- Horizontal Edges ---

# Top Row (Outer -> B=1)
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "B"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "C"}), {"R": 0, "B": 1}))

# Middle Row (Inner -> B=0)
hg.add_edge(Edge(EdgeType.E, frozenset({"D", "E"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"E", "F"}), {"R": 0, "B": 1}))

# Bottom Row (Outer -> B=1)
hg.add_edge(Edge(EdgeType.E, frozenset({"G", "H"}), {"R": 0, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"H", "I"}), {"R": 0, "B": 1}))


# --- Vertical Edges ---

# Left Column (Outer -> B=1)
hg.add_edge(Edge(EdgeType.E, frozenset({"A", "D"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"D", "G"}), {"R": 0, "B": 1}))

# Middle Column (Inner -> B=0)
hg.add_edge(Edge(EdgeType.E, frozenset({"B", "E"}), {"R": 1, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"E", "H"}), {"R": 0, "B": 1}))

# Right Column (Outer -> B=1)
hg.add_edge(Edge(EdgeType.E, frozenset({"C", "F"}), {"R": 0, "B": 1}))
hg.add_edge(Edge(EdgeType.E, frozenset({"F", "I"}), {"R": 0, "B": 1}))


# --- Square Faces (Q-edges) ---

# Top-Left (The "one small rectangle" with R=1)
hg.add_edge(Edge(EdgeType.Q, frozenset({"A", "B", "E", "D"}), {"R": 0}))

# Others (R=0)
hg.add_edge(Edge(EdgeType.Q, frozenset({"B", "C", "F", "E"}), {"R": 0}))
hg.add_edge(Edge(EdgeType.Q, frozenset({"D", "E", "H", "G"}), {"R": 0}))
hg.add_edge(Edge(EdgeType.Q, frozenset({"E", "F", "I", "H"}), {"R": 0}))


# --- Vertex Coordinates ---
hg.set_vertex_parameter("A", {"x": 0, "y": 0})
hg.set_vertex_parameter("B", {"x": 1, "y": 0})
hg.set_vertex_parameter("C", {"x": 2, "y": 0})
hg.set_vertex_parameter("D", {"x": 0, "y": 1})
hg.set_vertex_parameter("E", {"x": 1, "y": 1})
hg.set_vertex_parameter("F", {"x": 2, "y": 1})
hg.set_vertex_parameter("G", {"x": 0, "y": 2})
hg.set_vertex_parameter("H", {"x": 1, "y": 2})
hg.set_vertex_parameter("I", {"x": 2, "y": 2})
```

```python
hg.draw(use_positional_parameters=True)
```

```python
from hypergrammar.productions.prod_4 import Prod4
prod0 = Prod4()
new_hg = prod0.apply(hg)
if new_hg:
#     print(new_hg.get_edges())
    new_hg.draw(use_positional_parameters=True)
else:
    print("Production could not be applied.")
```
